## Eigenvalues and Eigenvectors Definition

Given a matrix $A$, and $A$ multiplies vector $x$, the role of matrix $A$ is like a function that transforming $x$ to $Ax$. $x$ is the input, $Ax$ is the output.   

We are interested in some vectors $x$ who points the same direction with $Ax$. Ths is the special case since $Ax$ points different directions with $x$ in most cases. In this special case, $Ax$ parallels to $x$, we call vector $x$ as the `eigen vector` of matrix $A$. Using mathematical langulage, it is presented as:

$$
A x=\lambda x
$$

where $\lambda$ is called `eigen value`. 

* If $\lambda$ is zero, then $Ax=0$. Thus in this case, the eigen vectors are in the null space of $A$.  

* If $A$ is singular, which means that $A$ transform some vector $x$ to zero vector. This means that $A$ has an eigen value $\lambda = 0$. 

* Then let's see the projection matrix $P=A\left(A^{T} A\right)^{-1} A^{T}$. Vector $b$ multiplies $P$ producing the projection vector $Pb$. In this case, $Pb$ and $b$ points the same direction iff when $b$ is in the column space of $A$, thus $Pb = 1\cdot b$. This dicates that all the vectors in the projection space is the eigen vector of projection matrix, and their corresonding eigen vectors are all 1. 
* Let's see the normal vector of projection space (column space of matrix $A$). Because $e \perp\operatorname{C}(A)$, thus $Pe=0e$. The eigen value of eigen vector $e$ is 0. 
* Another example, a 2 dimensional permutation matrix $A=\left[\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right]$. By multiplying the matrix, the elements of vector will exchange. Therefore,
    * The eigen vectors with eigen value 1 is like $\left[\begin{array}{l}1 \\ 1\end{array}\right]$
    * The eigen vectors with eigen value -1 is like $\left[\begin{array}{l}1 \\ -1\end{array}\right]$

Here is an important property of eigen values: for a $n\times n$ matrix, it has $n$ eigen values (may have same value), the sum of these eigenvalues is the same as the sum of diagnal elements of matrix. We call the sum of the diagnal elements as the `trace` of matrix. 

$$\sum_{i=1}^{n} \lambda_{i}=\sum_{i=1}^{n} a_{i i}$$

For the above permutation matrix $A=\left[\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right]$, we can see that $1+(-1) = 0+0$. This can help us easily calculate the eigen value. 

## Solve $Ax = \lambda x$

For $Ax = \lambda x$, there are two unknowns $x, \lambda$. We need some trick to solve it. We rewrite it as:

$$
(A-\lambda I) x=0
$$

We get the form that we have talked many times $Ax = 0$. If the above equation has solution, that means that $(A-\lambda I)$ must be singular matrix whose determinant is 0:


$$
\operatorname{det}(A-\lambda I) =0
$$

The above equation ,called as characteristic equation, has no $x$ now, which we can solve to get $\lambda$. After obtaining $\lambda$, we can substitute it to  $(A-\lambda I) x=0$ and get $x$. 

* Here is an example. $A=\left[\begin{array}{ll}3 & 1 \\ 1 & 3\end{array}\right]$. 
    * calculate $\operatorname{det}(A-\lambda I) =0 $$$\left|\begin{array}{cc}
3-\lambda & 1 \\
1 & 3-\lambda
\end{array}\right| =(3-\lambda)^{2}-1=\lambda^{2}-6 \lambda+8=0 \Longrightarrow  \lambda_{1}=4, \lambda_{2}=2$$
We can see that $\lambda_1 + \lambda_2 = 6$, which is exactly the sum of diagnal elements
    * Now we calclulate eigen vectors
    $$(A-4 I)x=0 \Longrightarrow \left[\begin{array}{cc}
-1 & 1 \\
1 & -1
\end{array}\right]x = 0 \Longrightarrow x_1=\left[\begin{array}{l}
1 \\
1
\end{array}\right]$$
Similarly, the another eigen vecotr is:

    $$(A-2 I)x=0 \Longrightarrow \left[\begin{array}{cc}
1 & 1 \\
1 & 1
\end{array}\right]x = 0 \Longrightarrow x_2=\left[\begin{array}{l}
1 \\
-1
\end{array}\right]$$
    * Note that for a symmetric, we will have real eigen values. For permutation matrix, we will have real eigen values and the eigen vectors are orthogonal. 
* Now let's see the relation of this example  $A=\left[\begin{array}{ll}3 & 1 \\ 1 & 3\end{array}\right]$ with the previous permutation example $A^{\prime}=\left[\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right]$. The eigen value and eigen vector of matrix $A^{\prime}$ is $\lambda_{1}=1, x_{1}=\left[\begin{array}{l}1 \\ 1\end{array}\right], \lambda_{2}=-1, x_{2}=\left[\begin{array}{c}-1 \\ 1\end{array}\right]$. It is easy to see that $A = A^{\prime} + 3I$. $A$ and $ A^{\prime}$ have the same eigen vectors and the difference their eigen values is 3. Why?
$$\begin{array}{l}
A x=\lambda x \Longrightarrow (A+3 I) x=\lambda x+3 x=(\lambda+3) x  \\
\end{array}$$
We can see that eigen vector $x$ reamins the same but $\lambda$ increases 3. 
    * Then can we say that: $A x=\lambda x, B x=\alpha x \Longrightarrow (A+B) x=(\lambda+\alpha) x$? If $B = 3I$, it is true. But if $B$ is any matrix, then it is not true because the eigen vector $x$ in the two equations may not be the same. The general case is $ A x=\lambda x, B y=\alpha y$, which we can not add them up. 

* Let's see a rotation matrix, $90^{\circ}$ rotation matrix $Q=\left[\begin{array}{cc}
\cos 90 & -\sin 90 \\
\sin 90 & \cos 90
\end{array}\right]=\left[\begin{array}{cc}
0 & -1 \\
1 & 0
\end{array}\right]$ (We denote it as $Q$ because rotation matrix is orthogonal matrix)
Previously,we have mentioned that the sum of eigen values is the trace. Here is another important property: the product of eigen values is determinant. 
$$
\prod_{i=1}^{n} \lambda_{i}=\operatorname{det} A
$$
    For matrix $Q$, we have $
\left\{\begin{array}{l}
\lambda_{1}+\lambda_{2}=0 \\
\lambda_{1} \cdot \lambda_{2}=1
\end{array}\right.
$. Now, let's think about does there exist some vectors that still parallels to itself after rotating $90^{\circ}$ ? No. This indicates that we can not find that eigen vector and eigen values for the roation matrix $Q$. 
Using the characteristic equation, we have:
$$
\operatorname{det}(Q-\lambda I)=\left|\begin{array}{cc}
\lambda & -1 \\
1 & \lambda
\end{array}\right|=\lambda^{2}+1=0 \Longrightarrow \lambda_{1}=i, \lambda_{2}=-i
$$ 
Note: In this example, $Q^{\top} = -Q$, it is the opposite of symmetric case and it has two pure complex eigen values. We mentioned before that symmetric matrix has real eigen values. These are the two extreems. 

* Let's see a even worse case, $A=\left[\begin{array}{ll}
3 & 1 \\
0 & 3
\end{array}\right]$, a triangular matrix. We can directly get its eigen values which is $3, 3$ because:
$$
\operatorname{det}(A-\lambda I)=\left|\begin{array}{cc}
3-\lambda & 1 \\
0 & 3-\lambda
\end{array}\right|=(3-\lambda)^{2}=0 \Longrightarrow \lambda_{1}=3, \lambda_{2}=3
$$
Let's compute its eigen vectors.
$$
(A-\lambda I) x=\left[\begin{array}{ll}
0 & 1 \\
0 & 0
\end{array}\right]\left[\begin{array}{l}
x_{1} \\
x_{2}
\end{array}\right]=\left[\begin{array}{l}
0 \\
0
\end{array}\right] \Longrightarrow  x_{1}=\left[\begin{array}{l}
1 \\
0
\end{array}\right]
$$
But when we substitute the second eigen value to the equation, we can not get a eigen vector that is independent with $x_1$. 
The reason we say this is worse case because it has two same eigen values and results one eigen vectors. In this case, we call matrix $A$ a `denegerate matrix`.
