This lecture focuses on the properties of `determinant`. We denote the determinant of $\operatorname{det}A$ as $|A|$. 

* Property 1: $\operatorname{det} I = 1$, the determinant of identity matrix is 1. 
* Property 2: Exchange rows reverses sign of determinant.  
    According to the first two properties, we can have the following for permutation matrix $P$: $$
\operatorname{det} P=\left\{\begin{array}{ll}
1 & \text { number of change is even } \\
-1 & \text { number of change is odd }
\end{array}\right.
$$
    Example:
 $$\left|\begin{array}{ll}1 & 0 \\ 0 & 1\end{array}\right|=1, \left|\begin{array}{ll}0 & 1 \\ 1 & 0\end{array}\right|=-1$$
    The determinant of a general $2\times 2$ matrix $\left|\begin{array}{ll}a & b \\ c & d\end{array}\right|$ is $ad - bc$

* Property 3: 
    * 3-a:  $\left|\begin{array}{ll}
t a & t b \\
c & d
\end{array}\right|=t\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right| $
    * 3-b: $\left|\begin{array}{cc}a+a^{\prime} & b+b^{\prime} \\ c & d\end{array}\right|=\left|\begin{array}{ll}a & b \\ c & d\end{array}\right|+\left|\begin{array}{cc}a^{\prime} & b^{\prime} \\ c & d\end{array}\right|$  
   It says that if all the other rows stay the same, the first row behaves like a linear function. We are saying linearity for each row. Not saying that $\operatorname{det} (A+B) = \operatorname{det} A + \operatorname{det} B$

* Property 4: If two rows are the same, the determinant are zero. We can use property 2 to prove this. 

* Property 5: Substracting $l$ times of $i$th row from $k$th row does not change the determinant. 
    Example:
    $$
\left|\begin{array}{cc}
a & b \\
c-l a & d-l b
\end{array}\right| \stackrel{3 . b}{=}\left|\begin{array}{c}
a & b \\
c & d
\end{array}\right|+\left|\begin{array}{cc}
a & b \\
-l a & -l b
\end{array}\right|\stackrel{3 . a}{=}\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right|-l\left|\begin{array}{ll}
a & b \\
a & b
\end{array}\right|\stackrel{4}{=}\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right|
$$

    Thanks to this property, we can first reduce the matrix and calculate the determinant more easily.  
    
* Property 6: If one row of a square matrix is 0, then its determinant is 0. Two ways to prove:
    * Use property 3-a: multipying $l$ to the 0 row, then $l \operatorname{det}A=\operatorname{det} A$ 
    * Add some row to the 0 row, then the two rows are the same and the determinant does not change (property 5). According to property 4, the determinant is 0. 
* Property 7: Upper matrix $\boldsymbol{U}=\left|\begin{array}{cccc}d_{1} & * & \cdots & * \\ 0 & d_{2} & \cdots & * \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & d_{n}\end{array}\right|$, then $\operatorname{det} U = d_1d_2\dots d_n$. 
    Prove: 
    $$\operatorname{det} \boldsymbol{U}=\left|\begin{array}{cccc}d_{1} & * & \cdots & * \\ 0 & d_{2} & \cdots & * \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & d_{n}\end{array}\right| \stackrel{Property 5} {\Longleftrightarrow}\operatorname{det}D=\left|\begin{array}{cccc}
d_{1} & 0 & \cdots & 0 \\
0 & d_{2} & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & d_{n}
\end{array}\right| \stackrel{Property 3} {\Longleftrightarrow} d_{n} d_{n-1} \cdots d_{1}\left|\begin{array}{cccc}
1 & 0 & \cdots & 0 \\
0 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & 1
\end{array}\right| =  d_1d_2\dots d_n $$

* Property 8: If $A$ is singular matrix, then $\operatorname{det} A = 0$. If and only if $A$ is invertible, $\operatorname{det} A \neq 0$. If $A$ is invertible, when $A$ is reduced to upper matrix, then each row has pivot, thus the determinant is the product of the pivot. Otherwise, the upper matrix will has some zero rows, thus the determinant is 0. 

    Example:
    $$
\left|\begin{array}{cc}
a & b \\
c & d
\end{array}\right| \stackrel{\text { elimination }}{\longrightarrow}\left|\begin{array}{cc}
a & b \\
0 & d-\frac{c}{a} b
\end{array}\right|=a d-b c
$$

* Property 9: $\operatorname{det} AB = (\operatorname{det} A)( \operatorname{det} B) $. Using this property, we can have:
    * $
\operatorname{det} I=\operatorname{det} A^{-1} A=\operatorname{det} A^{-1} \operatorname{det} A =  1 \Longrightarrow 
\operatorname{det} A^{-1}=\frac{1}{\operatorname{det} A}
$
    * $\operatorname{det} A^{2}=(\operatorname{det} A)^{2}$
    * $\operatorname{det} 2 A=2^{n} \operatorname{det} A$, suppose $A$ is $n \times n$.This is can be easily get by using property 3-a.  It's like the volume. If we double the side of a box, its volume becomes 8 times bigger.  

* Property 10: $\operatorname{det} A^{T}=\operatorname{det} A$.   
    Prove:
    $$\left|A^{T}\right|=|A| \rightarrow\left|U^{T} L^{T}\right|=|L U| \rightarrow\left|U^{T}\right|\underbrace{\left|L^{T}\right|}_{1}=\underbrace{|L|}_{1}|U|$$
    Thanks to this property, the above properties also apply to the column of matrix, e.g., exchanging two columns reverses the sign of determinant (Property 2)
