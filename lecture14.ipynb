In [lecture 10](https://nbviewer.jupyter.org/github/jswan95/Linear-Algebra-note/blob/main/lecture10.ipynb), we have mentioned that for a $m\times n$ matrix $A$ with rank $r$. Its row space $\operatorname{dimC}(A^\top)=r$ and null space $\operatorname{N}(A) = n-r$ are in $\mathbb{R}^n$ vector space. Its column space $\operatorname{dimC}(A)=r$ and left null space $\operatorname{dimN}(A^{\top}) = m-r$ are in $\mathbb{R}^m$

For vector $x, y$, if $x^{\top}\cdot y = 0 \Longrightarrow x_1y_1 + x_2y_2+\dots+x_ny_n=0$, we say $x, y$ are orthogonal. 

According to Pythagorean theorem, it says that the sides of a right triangle satisfies:

$$
\begin{aligned}
&\begin{aligned}
\|\vec{x}\|^{2}+\|\vec{y}\|^{2} &=\|\overrightarrow{x+y}\|^{2} \\
x^{T} x+y^{T} y &=(x+y)^{T}(x+y) \\
x^{T} x+y^{T} y &=x^{T} x+y^{T} y+x^{T} y+y^{T} x
\end{aligned}\\
&0=x^{T} y+y^{T} x \quad \text { for dot product, } x^{T} y=y^{T} x\\
&0=2 x^{T} y\\
&x^{T} y=0
\end{aligned}
$$

Therefore, we can say that the dot product of two orthogonal vectors are 0. Because the product of 0 vector with all vectors is 0, 0 vector is orthogonal to all vectors. 

Giving an example:

$$
\begin{aligned}
x=\left[\begin{array}{l}
1 \\
2 \\
3
\end{array}\right], y=\left[\begin{array}{c}
2 \\
-1 \\
0
\end{array}\right], x+y=\left[\begin{array}{l}
3 \\
1 \\
3
\end{array}\right]
\end{aligned}\\
\Longrightarrow |\vec{x}\|^{2}=14,\|\vec{y}\|^{2}=5,\|\overrightarrow{x+y}\|^{2}=19 \\
x^{T} y=1 \times 2+2 \times(-1)+3 \times 0=0
$$




Now we move to subspace. 

If subspace $S$ is orthogonal to subspace $T$, it means that every vector in $S$ is orthogonal to every vector in $T$. If two subspaces are orthogonal, their intersection will only contain 0 vector. 

Then how about the ortogonality among the four fundamental suspaces ?

For row space and null space:

The null space of $A$ is the solution space of $Ax = 0$. It can be written asï¼š

$$
\left[\begin{array}{c}
r o w_{1} \\
r o w_{2} \\
\vdots \\
r o w_{m}
\end{array}\right][x]=\left[\begin{array}{c}
0 \\
0 \\
\vdots \\
0
\end{array}\right]
$$

$$
\begin{aligned}
&\left[\mathrm{row}_{1}\right][x]=0\\
&\left[\mathrm{row}_{2}\right][x]=0\\
&\qquad \vdots\\
&\left[\mathrm{row}_{m}\right][x]=0
\end{aligned}
$$

The above equations tell that $x$ is orthogonal to each row of $A$. Next we show that $x$ is also orhtogonal to the linear combination of rows of $A$. 

$$
\left\{\begin{array}{l}
c_{1}\left(r o w_{1}\right)^{T} x=0 \\
c_{2}\left(r o w_{2}\right)^{T} x=0 \\
\vdots \\
c_{n}\left(r o w_{m}\right)^{T} x=0
\end{array}\right. \xrightarrow{summing up}
\left(c_{1} r o w_{1}+c_{2} r o w_{2}+\cdots+c_{n} r o w_{m}\right)^{T} x=0
$$

Therefore,we can say that row space and null space are two orthogonal subspace in $\mathbb{R}^n$. And it should also be noted that row space and null space couple into $\mathbb{R}^n$. 

Similarly, we can also show that column space and left null spaces are two orthogonal subspaces and they couple into $\mathbb{R}^m$. 

Giving an example:

For matrix $A=\left[\begin{array}{ccc}1 & 2 & 5 \\ 2 & 4 & 10\end{array}\right]\Longrightarrow m=2, n=3, \operatorname{rank}(A)=1, \operatorname{dimN}(A)=2$ 

Let $Ax = 0$, we obtain:

$$A x=\left[\begin{array}{ccc}1 & 2 & 5 \\ 2 & 4 & 10\end{array}\right]\left[\begin{array}{l}x_{1} \\ x_{2} \\ x_{3}\end{array}\right]=\left[\begin{array}{l}0 \\ 0\end{array}\right]\Longrightarrow x_{1}=\left[\begin{array}{c}
-2 \\
1 \\
0
\end{array}\right] \quad x_{2}=\left[\begin{array}{c}
-5 \\
0 \\
1
\end{array}\right]$$

A basis of the row space is $\left[\begin{array}{l}
1 \\
2 \\
5
\end{array}\right]$, which is obviously orthogonal to $x_1, x_2$. 

One more point should be noted is that we say the row space and the null space are `orthogonal complement` in $\mathbb{R}^n$, because row space (null space) contains $\underline{\text{all}}$ vectors that orthogonal to the null space (row space). 

This part is kind of preparation work for projection and linear regression. 

Suppose a $m\times n , m>n$ rectangular matrix $A$, and a linear equations $Ax = b$ without solution (b is not in the column space of A). 

This happens a lot in real world since the observed value $b$ main contain some noise. 

Our goal is to "solve" $Ax=b$. How to find the best possible solution? 

By introducing $A^{\top}A$, which is a $n\times n$ matrix, that is also a symmetric matrix:

$$
\left(A^{T} A\right)^{T}=A^{T}\left(A^{T}\right)^{T}=A^{T} A
$$

We can obtain a "good" equation by multiplying $A^{\top}$:

$$
Ax = b \Longrightarrow A^{T} A \hat{x}=A^{T} b
$$

This equation is solveable if $A^{T} A$ is invertible. 

Here are some facts:
* $N\left(A^{T} A\right) =N(A) $
* $\operatorname{rank}\left(A^{T} A\right) =\operatorname{rank}(A)$
* $A^{T} A$ is invertible if and if the null space of $A: \operatorname{N}(A) = \{0\}$, which means that the column vector of $A$ is linearly independent.
